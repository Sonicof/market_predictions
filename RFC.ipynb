```python
import yfinance as yf
```

```python
sp500=yf.Ticker("^GSPC")
```

```python
sp500=sp500.history(period="max")
```

```python
sp500
```

```python
sp500.plot.line(y="Close",use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tommorow"]=sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"]=(sp500["Tommorow"]>sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500=sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)
```

```python
train = sp500.iloc[:-100] #all except the last 100
test = sp500.iloc[-100:] #last 100 will be used for testing the model

predictors = ["Close","Volume","Open","High","Low"]

model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds
```

```python
import pandas as pd
preds=pd.Series(preds,index=test.index,name="Predictions")
```

```python
preds
```

```python
precision_score(test["Target"],preds)
```

```python
combined= pd.concat([test["Target"], preds], axis=1)
combined
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds,index=test.index, name="Predictions")
    combined= pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data,model,predictors,start=2500,step=250):  #every training year is of 250 days...we train one model on 10 years of data and then keep on training it every year wiht a step value of 250 days based on our growing model
    all_predictions=[]
    for i in range(start,data.shape[0],step):
        train = data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions=predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)    
```

```python
predictions= backtest(sp500,model,predictors)
predictions
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
horizons = [2,5,60,250,1000]
new_predictors=[]
for horizon in horizons:
    rolling_average = sp500.rolling(horizon).mean()
    ratio_column=f"Close_Ratio_{horizon}"
    sp500[ratio_column]= sp500["Close"] / rolling_average["Close"]
    trend_column = f"Trend_{horizon}"
    sp500[trend_column]= sp500.shift(1).rolling(horizon).sum()["Target"] #shifting the trend column by 1 so that it is the sum of the last two days excluding the current day otherwise there will be leakage and would not work in real world
    new_predictors +=[ratio_column,trend_column]
    
```

```python
sp500
```

```python
sp500 = sp500.dropna()
```

```python
model=RandomForestClassifier(n_estimators=200,min_samples_split=50,random_state=1)
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict_proba(test[predictors])[:,1]  #returns a probability wheather a row will be zero or one
    preds[preds>=.6]=1
    preds[preds<.6]=0 
    preds=pd.Series(preds,index=test.index, name="Predictions")
    combined= pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions=backtest(sp500,model,new_predictors)
```
